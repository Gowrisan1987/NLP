```
path="/content/drive/My Drive/Hackathon/amazonreviews.tsv"
rev=pd.read_csv(path,sep='\t')
```

```
sa.polarity_scores(rev.loc[0]['review'])
```

```
rev['scores']=rev['review'].apply(lambda x: sa.polarity_scores(x))
```

```
rev.head()
```

```
rev.loc[0]['scores']['compound']
```

```
rev['compund']=rev['scores'].apply(lambda score_dict: score_dict['compound'])
```

```
rev.head()
```

```
rev['predict']=rev['compund'].apply(lambda c: 'pos' if c>0 else 'neg')
```

```
rev.head()
```

```
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```
val=accuracy_score(rev['label'],rev['predict'])
```

```
val
```

```
print(confusion_matrix(rev['label'],rev['predict']))
```

```
print(classification_report(rev['label'],rev['predict']))
```
